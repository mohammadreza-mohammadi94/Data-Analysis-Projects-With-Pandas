# 0062
# Loading Data from DataFrame into SQLlite Database

```python
import pandas as pd
import json
import sqlite3 as sq3
```

```python
with open('some_movies.json') as f:
    data = json.load(f)
```

```python
data
```

```python
df = pd.json_normalize(data, sep= "_")
df
```

```python
movies = df[['id', 'title', 'revenue', 'budget', 'belongs_to_collection_name', 'release_date']].copy()
```

```python
movies
```

```python
movies.info()
```

***Release_date must be datetime object***

```python
movies['release_date'] = pd.to_datetime(df['release_date'])
```

creating a new DataFrame called movies by selecting specific columns from df using the square bracket notation ([]) and the copy() method. Let me explain this code to you.

The [['id', 'title', 'revenue', 'budget', 'belongs_to_collection_name', 'release_date']] syntax is used to select specific columns from the df DataFrame. Each column name is enclosed in quotes and separated by a comma.

The resulting output will be a new DataFrame called movies, which contains only the selected columns from df.

You're also using the copy() method to create a copy of the selected columns, rather than a reference to the original DataFrame. This can be useful for modifying the new DataFrame without affecting the original DataFrame.

```python
movies.info()
```

```python
movies['revenue'] = movies['revenue']/1000000
movies['budget'] = movies['budget'] / 1000000
```

```python
movies.info()
```

```python
movies.head(3)
```

```python
votes = df[['id', 'vote_count', 'vote_average']].copy()
```

```python
votes
```

```python
genres = pd.json_normalize(data = data, record_path= "genres", meta= 'id', record_prefix='genre_')
```

```python
genres
```

```python
prod = pd.json_normalize(data = data , record_path='production_companies', meta = 'id', record_prefix='comp_')
prod
```

```python
con = sq3.connect('movies.db')
```

```python
con
```

```python
movies.to_sql('Movies', con, index = False)
```

```python
votes.to_sql('Votes', con, index = False)
```

```python
genres.to_sql('Genres', con, index = False)
```

```python
prod.to_sql('Prod', con, index = False)
```

```python
con.execute('SELECT * FROM sqlite_master').fetchall()
```

```python
con.execute("SELECT name FROM sqlite_master WHERE type = 'table' ORDER BY name").fetchall()
```

```python
con.close()
```

# 0066
# Join Queries

```python
import pandas as pd 
import sqlite3 as sq3
```

```python
con = sq3.connect('movies.db')
```

```python
pd.read_sql("SELECT * FROM Movies", con)
```

```python
pd.read_sql("SELECT * FROM Votes", con)
```

```python
pd.read_sql("SELECT * \
            FROM Movies \
            JOIN Votes \
            ON Movies.id = Votes.id", con)
```

```python
pd.read_sql("SELECT * \
            FROM Movies \
            JOIN Votes \
            ON Movies.id = Votes.id", con).head(3)
```

```python
pd.read_sql("""SELECT Movies.id, Movies.title, Votes.vote_average 
            FROM Movies 
            JOIN Votes 
            ON Movies.id = Votes.id""", con, index_col= 'id')
```

```python
pd.read_sql("""SELECT Movies.id, Movies.title, Votes.vote_average
               FROM Movies
               JOIN Votes
               ON Movies.id = Votes.id
               WHERE Votes.vote_average >8""", con, index_col='id')
```

```python
pd.read_sql("""SELECT Movies.id, Movies.title, Votes.vote_average
                FROM Movies
                JOIN Votes
                ON Movies.id = Votes.id
                WHERE Votes.vote_average > 8
                ORDER BY Movies.budget ASC""", con, index_col='id')
```

# 0067
# Final case Study

```python
import pandas as pd
import sqlite3 as sq3
```

```python
con = sq3.connect('movies.db')
```

```python
pd.read_sql("SELECT * FROM Prod", con)
```

```python
df = pd.read_sql("""SELECT Prod.id , Prod.comp_name, Movies.revenue, Movies.title
                    FROM Prod
                    LEFT JOIN Movies
                    ON Prod.id = Movies.id""", con)
```

```python
df
```

```python
df.groupby('comp_name')['revenue'].sum().sort_values(ascending = True)
```

```python
pd.read_sql("""SELECT Prod.comp_name
                FROM Prod
                LEFT JOIN Movies
                ON Prod.id = Movies.id
                WHERE Movies.title = 'Titanic'""", con)
```

```python
df2 = pd.read_sql("""SELECT Genres.id, Genres.genre_name , Movies.revenue, Movies.title
                    FROM Genres
                    LEFT JOIN Movies
                    ON Genres.id = Movies.id
                    """, con)
```

```python
df2
```

```python
df2.groupby("genre_name")['revenue'].sum().sort_values(ascending = True)
```

```python
pd.read_sql("""SELECT Genres.genre_name
                FROM Genres
                LEFT JOIN Movies
                ON Genres.id = Movies.id
                WHERE Movies.title = 'Frozen II'""", con)
```

```python
con.close()
```
