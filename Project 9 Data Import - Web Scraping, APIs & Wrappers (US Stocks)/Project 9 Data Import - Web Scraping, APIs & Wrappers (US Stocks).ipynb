```python
import unicodedata
```

The unicodedata module has a number of functions that can be used to work with Unicode characters. Some of the most commonly used functions include:

- lookup(): This function looks up a Unicode character by name and returns its corresponding code point.
- name(): This function returns the name of a Unicode character given its code point.
- category(): This function returns the category of a Unicode character.
- combining(): This function returns the combining class of a Unicode character.
- decomposition(): This function returns the decomposition of a Unicode character.

The unicodedata module is a powerful tool for working with Unicode data. It can be used to perform a variety of tasks, such as:

- Detecting and handling special characters
- Normalizing Unicode text
- Converting between different encodings
- Analyzing the structure of Unicode text

```python
const['Symbol'].apply(lambda x: unicodedata.normalize("NFKD", x))
```

The `apply()` function is a Pandas function that can be used to apply a function to each row or column of a DataFrame. In this case, the apply() function is used to apply the `unicodedata.normalize("NFKD")` function to each string in the Symbol column.

The output of the code will be a new DataFrame with the Symbol column normalized.

Here is a more detailed explanation of each function in the code:

- const: This is the DataFrame that contains the data.
- Symbol: This is the column in the DataFrame that contains the strings that will be normalized.
- lambda x: unicodedata.normalize("NFKD", x): This is a lambda function that takes a string as input and returns the normalized string.
- apply(): This function applies the lambda function to each string in the Symbol column.

```python
const['Symbol'] = const['Symbol'].apply(lambda x: unicodedata.normalize("NFKD", x))
```

***Valid values for form are 'NFC', 'NFKC', 'NFD', and 'NFKD'.***

```python
const['Symbol'][1]
```

```python
const['Ticker'] = const['Symbol'].str.split(": ").apply(lambda x: x[-1])
```

he str.split() function is a Pandas function that can be used to split a string into a list of strings based on a delimiter. In this case, the str.split() function is used to split the strings in the Symbol column using the : delimiter.

The apply() function is a Pandas function that can be used to apply a function to each row or column of a DataFrame. In this case, the apply() function is used to apply the lambda function to each list of strings in the Symbol column.

The lambda function is a Python function that takes a list of strings as input and returns the last element in the list.

The output of the code will be a new DataFrame with the Symbol column containing the last element of each list of strings.
